```python
import pandas as pd
import matplotlib.pyplot as plt


import numpy as np
```

```python
df=pd.read_csv('salary_dataset.csv')
```

```python
df
```

```python
df.sort_values(by='Salary')
```

```python
df.plot.scatter(x='YearsExperience', y='Salary', c='red', colorbar=True)
plt.title('Experience vs Salary', fontsize=15)
plt.xlabel('Salary')
plt.ylabel('Experience',)
plt.legend('Hot map', fontsize=15)
plt.show()
```

```python
df.plot.hist(x='Salary', y='YearsExperience')
plt.xlabel('Salary')
plt.ylabel('Experience')
plt.title('Salary VS Experience')
plt.show()
```

```python
df.plot(x='Salary', y='YearsExperience')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression 
X = df[['YearsExperience']]
Y = df['Salary']

model = LinearRegression()  
model.fit(X, Y)
```

```python
y_pred = model.predict(X)

plt.figure(figsize=(10, 6))
plt.scatter(X, Y, label='Actual data')
plt.plot(X, y_pred, color='red', label='Predictions')
plt.legend()
plt.show()
```

```python
import numpy as np

Exp = 10.6
Exp_reshaped = np.array([Exp]).reshape(-1, 1)
Sal = model.predict(Exp_reshaped)

print('Years of Experience=', Exp)
print('Expected Salary=', Sal[0])  # Use [0] to get the scalar prediction value
```
